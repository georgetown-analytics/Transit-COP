```python
# In order to match the dataset from carsoup.com, we need to fix the columns car_vendor and car_location
# to show at least a city nme and a state. Before doing that and in order to do loops based on strings,
# car_location and car_vendor must be converted to a string type they are set up per default to dtype object
```

```python
np.unique(carsdotcom['car_location'].values)
```

```python
#Virginia = Alexandria, Arlington, Falls Church 
#Maryland = Bethesda, Bladensburg, Camp Springs, Capitol Heights, Chevy Chase, College Park, District Heights,
#Marlow Heights,  Hyattsville, Oxon Hill, Riverdale, Silver Spring, Suitland, Takoma Park, Temple Hill, 
#Temple Hills
```

```python
carsdotcom['car_location'] = carsdotcom['car_location'].astype('str')
carsdotcom['car_vendor'] = carsdotcom['car_vendor'].astype('str')

for index, row in carsdotcom.iterrows():
    
    if row['car_vendor'] == 'Washington':
        temp_car_location_DC = 'DC'
        row['car_location'] = temp_car_location_DC
        carsdotcom.ix[index]['car_location'] = row['car_location']
        #print carsdotcom.ix[index]['car_location']
            
    elif row['car_vendor'] == 'Alexandria' or  row['car_vendor'] =='Arlington' or row['car_vendor']== 'Falls Church':
        temp_car_location_VA = 'VA'
        row['car_location'] = temp_car_location_VA
        carsdotcom.ix[index]['car_location'] = row['car_location']
        #print row['car_location']
        #print carsdotcom.ix[index]['car_location']
   
    else :
        temp_car_location_MD = 'MD'
        row['car_location'] = temp_car_location_MD
        carsdotcom.ix[index]['car_location'] = row['car_location']
        #print row['car_location']
        #print carsdotcom.ix[index]['car_location']
```
