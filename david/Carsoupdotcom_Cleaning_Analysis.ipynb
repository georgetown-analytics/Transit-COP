```python
# First of all, since columns car_vendor, car_location, car_detail_link, car_img_link and car_engine are swapped,
# we need to allocate the valaues to the right columns.
#Column car_vendor is empty however it should contain the values in car_location, the values in car_deatil_link should
# be placed in car_location and so on so forth
```

### Moving missplaced values to their right column

```python
car_vendor = cars['car_vendor']
car_location = cars['car_location']
car_detail_link = cars['car_detail_link']
car_img_link = cars['car_img_link']
car_engine = cars['car_engine']
column_18 = cars['column_18']

temp_car_location = car_location
temp_car_detail_link = car_detail_link
temp_car_img_link = car_img_link
temp_car_engine = car_engine
temp_column_18 = column_18
```

```python
cars['car_vendor'] = temp_car_location
cars['car_location'] = temp_car_detail_link
cars['car_detail_link'] = temp_car_img_link
cars['car_img_link'] = temp_car_engine 
cars['car_engine'] = temp_column_18 
```

```python
cars.head()
```

```python
# We are under the impression that there will be a lot duplicated rows since many sellers usually post adds more than
# once in order to increase the probability of selling and visibility. Let's see if we have duplicated rows

cars.duplicated().sum()
```

```python
# Surprisingly we have no duplicated rows.........????????
# A row is considered duplicate when all the column values are the same, however there is a unique value for each 
# column i.e. car_id. Although the seller is posting many times the same add, the website provides a different car_id
# each time. 

# Under those premises we won't be able to check if any duplicate unless we don't take the car_id column into account,
# let's get rid of the car_id and check duplicates again
```

```python
cars_fix = cars.ix[:, {3, 4, 6, 8, 9, 11, 12, 13,14, 17}]
cars_fix.head()
```

```python
cars_fix.duplicated().sum()
```

```python
# Bingo! let's take a look on those 24 duplicated rows
cars_fix.ix[cars_fix.duplicated(keep=False)]
```

```python
# Here is the problem, we don't want 24 rows duplicated but we need the car_id in our dataframe because for further 
# analysis we'll need to match the cars with the image and website links (from our primary dataframe) and the car_id is
# the only column which will do that match for us.

# Let's add again the car_id to our datframe and get rid of the rows which are showing duplicated values for columns
# 1 to column 10 (keeping the last value/row)

cars_fix = cars.ix[:, {1, 3, 4, 6, 8, 9, 11, 12, 13,14, 17}]
cars_fix.head()
```

```python
cars.columns
```

```python
# Let's get rid of the duplicated rows from column 1 to 10 and keep at least one of the duplicated rows (the last)

df = cars_fix.drop_duplicates(['car_make', 'car_model', 'car_year', 'car_mileage', 'car_price', 'car_transmission',
                          'car_driveTrain', 'car_vendor', 'car_location', 'car_engine'], keep = 'first')
```

```python
df
```

```python
# After droping the duplicated rows we have a dataframe amounted of 17,043 counts which is in agreement with the 
# initial dataframe about 17,067 minus 24 duplicates = 17,043. Let's move forward!
```

### Nan values 2nd

```python
df.isnull().sum()
```

```python
# There are still null values to fill, let's make some hypothesis:

# 1.car_transmission empty values will be filled with value: 'automatic' which is most common value for that variable
# 2.car_driveTrain empty values will be filled with 'FWD' which is also the most common value for drivetrain 
# 3. here are 336 car engine empty values because there were no records in the website, those won't be filled!
```

```python
df['car_transmission'].fillna('Automatic')
df['car_driveTrain'].fillna('FWD')
df.head(10)
```

```python
df.isnull().sum()
```

```python
# We are good ! the only Nan values are the ones from car_engine column which are also empty in the website
```
