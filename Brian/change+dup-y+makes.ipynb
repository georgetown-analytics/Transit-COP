```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('/home/brian/Documents/project/data/most_recent_version.csv', 
            index_col = 0 , error_bad_lines = False)
```

We see that some car_make values are not technically duplicates, but actually are. For example, Aston Martin and Aston-Martin. 

```python
df.groupby('car_make').count()
```

Let's replace them...

```python
df = df.replace('Aston-Martin', 'Aston Martin')
```

```python
df = df.replace('INFINITI', 'Infiniti')
```

```python
df = df.replace('smart', 'Smart')
```

```python
df = df.replace('TESLA', 'Tesla')
```

```python
df = df.replace('RAM', 'Ram')
```

```python
df = df.replace('Land-Rover', 'Land Rover')
```

...and see our results!

```python
df.groupby('car_make').count()
```

Now to save the new .csv

```python
df.to_csv("most_recent_version2.csv", index=False)
```
